```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go

import pandas as pd

data = pd.read_csv("datadummy.csv")
site_lat = data.lat
site_lon = data.long
rsrp = data.RSRP
rsrq = data.RSRQ

print(site_lat,site_lon)
```

```python
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Extract the 'rsrp' and 'rsrq' columns from the DataFrame
# Create the scatter plot
plt.plot(site_lat,rsrq)

# Set labels and title
plt.xlabel('RSRP')
plt.ylabel('RSRQ')
plt.title('Scatter Plot of RSRP vs RSRQ')

# Show the plot
plt.grid(True)
plt.show()
```

```python
from ipyleaflet import Map, Marker, basemaps, basemap_to_tiles
m = Map(
    basemap=basemap_to_tiles(basemaps.CartoDB.Positron),
    center=(-6.958979, 107.623104),
    zoom=10
)
for index, row in data.iterrows():
    marker = Marker(location=(row['lat'], row['long']))
    m.add_layer(marker)


m.add_layer(Marker(location=(-6.958979, 107.623104)))
m.add_layer(Marker(location=(-6.959846,107.623070)))

m
```
