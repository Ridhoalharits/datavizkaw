```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, ListedColormap

data = pd.read_csv('datadummy2.csv')

site_long = data.lon
site_lat = data.lat
data_array = np.array(data.RSRQ)
print(data_array)

```

```python
import folium

m = folium.Map(location=(-6.958979, 107.623104),zoom_start=15)
for index, row in data.iterrows():
    folium.Marker(
    location=(row['lat'], row['lon']),
    tooltip=row['RSRQ'],
    popup="Mt. Hood Meadows",
    icon=folium.Icon(icon="signal"),
).add_to(m)
    
m
```

```python
m = folium.Map(location=(-6.958979, 107.623104),zoom_start=15)
lines = []
for index, row in data.iterrows():
    coordinates = [row['lat'], row['lon']]
    lines.append(coordinates)


for index, row in data.iterrows():
    folium.PolyLine(
         locations=lines,
        color='blue'
    ).add_to(m)

# Display the map in the Jupyter Notebook
m
```

```python
fig, (ax1, ax2,ax3) = plt.subplots(3, 1,figsize=(12,9))
fig.suptitle('Parameter terhadap jarak')

ax1.plot(data.no,data.RSRQ)
ax1.set_xlabel('Jarak (m)')
ax1.set_ylabel('RSRQ')
ax2.plot(data.no,data.RSSNR)
ax2.set_xlabel('Jarak (m)')
ax2.set_ylabel('RSSNR')
ax3.plot(data.no,data.RSRP)
ax3.set_xlabel('Jarak (m)')
ax3.set_ylabel('RSRP')
#plt.xticks(data.no)
plt.show()
```
